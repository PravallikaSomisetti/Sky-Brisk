# **2. HANDS-ON PRACTICE (Basic Visualization)**
## Matplotlib Example

```python
import matplotlib.pyplot as plt

marks = [65, 70, 75, 80, 85, 90]

plt.plot(marks)
plt.title("Student Marks Trend")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()
```

## Seaborn Example

```python
import seaborn as sns
import matplotlib.pyplot as plt

scores = [60, 65, 70, 75, 80, 85, 90]

sns.histplot(scores, kde=True)
plt.title("Distribution of Scores")
plt.show()
```

# **Client Project**

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load a Dataset

```python
df = pd.read_csv("/content/student_performance.csv.txt")
print(df)
```

## 3. Scatter polt

```python
plt.figure()
plt.scatter(df["study_hours"], df["marks"])
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

## 4. Histogram

```python
plt.figure()
plt.hist(df["marks"], bins=5)
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

## 5. Pair plot

```python
sns.pairplot(df)
plt.show()
```

## Heatmap (Correlation)

```python
plt.figure()
correlation = df.corr()
sns.heatmap(correlation, annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```
