# **Theory**

# **1. Hands-On (Data Transformations) using lists, lambda, comprehension**
*Data Transformation means modifying data into a required format using operations or functions.*


```python
# Real-world data: Student marks
marks = [45, 67, 89, 45, 90, 34, 67]
```

```python
# Remove duplicates
unique_marks = list(set(marks))
print("Unique Marks:", unique_marks)
```

```python
# Square of marks using list comprehension
squared_marks = [m**2 for m in unique_marks]
print("Squared Marks:", squared_marks)
```

```python
# Filter passed students (marks >= 40)
passed_students = list(filter(lambda x: x >= 40, unique_marks))
print("Passed Students Marks:", passed_students)
```

```python
# Sum of squares
sum_of_squares = sum(squared_marks)
print("Sum of Squares:", sum_of_squares)
```

# **2. Functions, Lambda & Recursion**

```python
# Function to calculate square
def square(n):
    return n * n
print("Square of 6:", square(6))
```

```python
# Function to calculate average
def calculate_average(data):
    return sum(data) / len(data)

scores = [78, 85, 90, 88]
print("Average Score:", calculate_average(scores))
```

```python
# Lambda function to add two numbers
add = lambda a, b: a + b
print("Addition:", add(5, 3))
```

```python
# Lambda function to calculate discount
discount_price = lambda price: price - (price * 0.10)
print("Price after discount:", discount_price(1000))
```

```python
# Recursive function to find factorial
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
print("Factorial of 5:", factorial(5))
```

```python
# Recursive function to calculate factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial of 6:", factorial(6))
```

# **Client Project: Data Cleaning Script**
*Data Cleaning: The process of removing duplicates, filtering incorrect data, and improving data quality.*

**Problem:**

Remove duplicate values

Filter valid data

Clean dataset using Python data structures

```python
# Real-world data: Sensor temperature readings (in °C)
temperature_readings = [25, 28, 30, -5, 28, 35, 40, 30, -3, 45]

print("Raw Temperature Data:", temperature_readings)

# Step 1: Remove duplicate values
unique_temperatures = list(set(temperature_readings))

# Step 2: Filter valid temperature values (0°C to 50°C)
valid_temperatures = [t for t in unique_temperatures if 0 <= t <= 50]

print("Cleaned Temperature Data:", valid_temperatures)

# Step 3: Calculate statistics
average_temp = sum(valid_temperatures) / len(valid_temperatures)
max_temp = max(valid_temperatures)
min_temp = min(valid_temperatures)

print("Average Temperature:", average_temp)
print("Maximum Temperature:", max_temp)
print("Minimum Temperature:", min_temp)

```
